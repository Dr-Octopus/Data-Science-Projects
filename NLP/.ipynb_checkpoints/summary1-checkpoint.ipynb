# BBC Articles Summarization and Topic Modelling

Hello!
This is a simple example of Natural Language Processing (NLP) summarization and topic modelling. The code has three main parts as explained below.
## 1. Gather Articles
The first part uses beautifulsoup to extracts all the news articles currently on the BBC news website in the following section: (world, business, UK, technology, science and environment, stories, entertainment and arts, and health). A unique set is then acquired to remove any articles that appear in multiple sections, and any blog pages or in-pictures pages are also dropped. The title, article body, and date are then extracted from each page only if it is available, this removes any video articles.
## 2. Article Summary
The article summary is generated based on a simple extractive method that ranks the sentences based on the words with the highest frequency. The main algorithm was taken from [https://stackabuse.com/text-summarization-with-nltk-in-python/](https://stackabuse.com/text-summarization-with-nltk-in-python/). 
## 3. Topic Modelling Using Latent Dirichlet Allocation (**LDA**)
LDA topic modeling is used on all the articles to generate 10 global topics. LDA is also used on each single article to generate a local topic. Top 10 words of the local and global topics are generated for each article.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import nltk
from tqdm import tqdm
from IPython.display import display, Markdown, display_markdown, HTML
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction import text
from datetime import datetime
from datetime import timedelta
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
sns.set()
from wordcloud import WordCloud, STOPWORDS
```

```python
all_cats = ['world', 'business', 'uk', 'technology', 'science_and_environment', 'stories', 'entertainment_and_arts', 'health']
# cat = all_cats[-1]#.split('_')[0]
url = 'https://www.bbc.com/news/'
links_only = []

for cat in all_cats:
#     print(cat)
    data = requests.get(url+cat)
    soup = BeautifulSoup(data.content, 'html.parser')

    all_links = soup.find_all('a', class_='gs-c-promo-heading gs-o-faux-block-link__overlay-link gel-pica-bold nw-o-link-split__anchor')
    links_only.extend(link['href'] for link in all_links 
                  if link['href'].split('/')[1] == 'news')
# links_only
```

```python
links_only = list(set(links_only))
```

```python
# clean by removing blogs and in-pictures
links_only = [link for link in links_only if link.split('/')[2].split('-')[0] not in ('blogs', 'in')]
```

```python
def build_dataset(links_only):
    news_data = []
    base_url = 'https://www.bbc.com'
    for url in tqdm(links_only):
        news_category = url.split('/')[2].split('-')[0]
        data = requests.get(base_url+url)
        soup = BeautifulSoup(data.content, 'html.parser')
        
        body = soup.find('div', class_='story-body__inner')
        if body:
            body = body.find_all('p')
            p_text = [p.get_text() for p in body]

            article_text = ""
            for p in p_text:
                article_text += p

            news_articles = [{'news_headline': soup.find('h1', class_='story-body__h1').string if soup.find('h1', class_='story-body__h1') else '',
                              'news_article': article_text,
                              'news_category': news_category,
                              'news_date': soup.find('div', class_='date date--v2').string,
                              'news_link': base_url+url}
                            ]
            news_data.extend(news_articles)
        
    df =  pd.DataFrame(news_data)
    df = df[['news_headline', 'news_article', 'news_date', 'news_category', 'news_link']]
    return df
```

```python
df = build_dataset(links_only)
```

```python
print(f'Found {len(df)} articles with text.')
```

```python
# Sort by date and remove any articles older than 1 week
df["news_date"] = df["news_date"].apply(lambda x: datetime.strptime(x, '%d %B %Y'))
df.sort_values(by=["news_date"], inplace=True, ascending=False)
df.drop(df[df["news_date"]<df["news_date"].iloc[0]+timedelta(days=-7)].index, inplace=True)

print(f'Found {len(df)} articles with text.')
```

```python
def get_summary(article_text, num_sents):
    # Removing Square Brackets and Extra Spaces
    article_text = re.sub(r'\[[0-9]*\]', ' ', article_text)
    article_text = re.sub(r'\s+', ' ', article_text)

    # Removing special characters and digits
    formatted_article_text = re.sub('[^a-zA-Z]', ' ', article_text )
    formatted_article_text = re.sub(r'\s+', ' ', formatted_article_text)

    # Create a list of sentences
    sentence_list = nltk.sent_tokenize(article_text)

    # Import stopwords
    stopwords = nltk.corpus.stopwords.words('english')

    # Find word frequancies
    word_frequencies = {}
    for word in nltk.word_tokenize(formatted_article_text):
        if word not in stopwords:
            if word not in word_frequencies.keys():
                word_frequencies[word] = 1
            else:
                word_frequencies[word] += 1

    # Find maximum word frequancy
    maximum_frequncy = max(word_frequencies.values())

    for word in word_frequencies.keys():
        word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)

    # Assign sentence scores
    sentence_scores = {}
    for sent in sentence_list:
        for word in nltk.word_tokenize(sent.lower()):
            if word in word_frequencies.keys():
                if len(sent.split(' ')) < 30:
                    if sent not in sentence_scores.keys():
                        sentence_scores[sent] = word_frequencies[word]
                    else:
                        sentence_scores[sent] += word_frequencies[word]

    # Print out summary
    import heapq
    summary_sentences = heapq.nlargest(num_sents, sentence_scores, key=sentence_scores.get)

    summary = ' '.join(summary_sentences)
    
    return summary
```

```python
list_summaries = [get_summary(df['news_article'].iloc[j], 5) for j in range(len(df))]
```

```python
df['summary'] = list_summaries
# df
```

```python
my_stop_words = text.ENGLISH_STOP_WORDS.union(['said','say','says','mr','ms','mrs','told','just'])

def get_local_topic():
    article_topic = []
    for i in range(len(df["news_article"])):
        cv = CountVectorizer(stop_words=my_stop_words)

        dtm = cv.fit_transform(df["news_article"].iloc[i:i+1])

        LDA = LatentDirichletAllocation(n_components=1,random_state=42)
        LDA.fit(dtm)

        topic_words = []
        num_words = 10
        for index, topic in enumerate(LDA.components_):
            words = [cv.get_feature_names()[wi] for wi in topic.argsort()[-1:-num_words-1:-1]]
            topic_words.append(words)
        article_topic.extend(topic_words)
    return article_topic
article_topic = get_local_topic()
```

```python
# Global topic assuming 10 topics
cv = CountVectorizer(stop_words=my_stop_words)
dtm = cv.fit_transform(df["news_article"])

LDA = LatentDirichletAllocation(n_components=10,random_state=42)
LDA.fit(dtm)

topic_words = []
num_words = 10
for index, topic in enumerate(LDA.components_):
    words = [cv.get_feature_names()[wi] for wi in topic.argsort()[-1:-num_words-1:-1]]
#     print(f'Top {num_words} words for topic #{index}')
#     print(words)
#     print('\n')
    topic_words.append(words)
    
topic_result = LDA.transform(dtm)
global_topic_details = []
for i in topic_result.argmax(axis=1):
    global_topic_details.append(topic_words[i])
```

```python
df['Local_Topic_details'] = article_topic
df['Global_Topic_details'] = global_topic_details
# df.head()
```

```python
# PLot the top 30 most frequent words in all of the articles
vocab = cv.get_feature_names(); 
count = dtm.toarray().sum(axis=0)
df_dict = pd.DataFrame({"vocab": vocab, "count": count})
df_dict.set_index(keys="vocab", inplace=True)
df_dict.sort_values(by="count",ascending=False, inplace=True)
top_vocab = df_dict.head(30)
```

```python
fig, ax = plt.subplots(figsize=(15,5))
sns.barplot(ax=ax, x=top_vocab.index, y=top_vocab["count"], palette="rocket").set_title('Top 30 most frequent words in all of the articles')
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(10,10))
wc = WordCloud(background_color="black", max_words=1000, 
               stopwords=STOPWORDS, max_font_size= 40, width=400, height=400, scale=5)
wc.generate(' '.join(list(df["news_article"])))
plt.title("Word Cloud of all articles", fontsize=20)
plt.imshow(wc.recolor(colormap= 'Pastel1_r'))
plt.axis('off')
plt.tight_layout(pad=0)
```
