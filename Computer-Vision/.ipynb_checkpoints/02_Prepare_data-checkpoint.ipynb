```python
import cv2
import glob
import pandas as pd
import numpy as np
from imutils import build_montages
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# import real and fake filenames
files_real = glob.glob('Image_samples/*_r_*.png')
files_fake = glob.glob('Image_samples/*_f_*.png')
print('Real Images:', len(files_real))
print('Fake Images:', len(files_fake))
```

```python
# Randomly sample 9 pairs of images and generate mask to visualize
idx = np.random.randint(low=0, high=len(files_real), size=9)
imgs = []
for i in idx:
    # real image
    img_real = cv2.imread(files_real[i])
    # fake image
    img_fake = cv2.imread(files_fake[i])
    # generate mask if difference between images is greater than 10 
    img_diff = cv2.subtract(img_real,img_fake)
    # if greater than threshold then set to white (255) (only to visualize)
    img_diff = np.where(img_diff>10,255,0).astype(dtype=np.uint8)
    # collapse the three channels
    img_diff = np.max(img_diff,axis=-1)
    # expand back to three channel (only to visualize)
    img_diff = img_diff[:, :, None] * np.ones(3, dtype=int)[None, None, :]

    imgs.extend([img_real,img_fake,img_diff])
    
montage = build_montages(imgs, (224, 224), (3, 9))[0]
fig, ax = plt.subplots(figsize=(7.5, 22.5))
ax.imshow(cv2.cvtColor(montage, cv2.COLOR_BGR2RGB))
plt.tight_layout()
plt.axis('off');
```

```python
# Create the actual masks
path_fake = 'Seg_samples/'
path_real = 'Seg_samples/'
for i in range(len(files_real)):
    # real image
    img_real = cv2.imread(files_real[i])
    # fake image
    img_fake = cv2.imread(files_fake[i])
    # generate mask if difference between images is greater than 10 
    img_diff = cv2.subtract(img_real,img_fake)
    # if greater than threshold then set to 1
    img_diff = np.where(img_diff>10,1,0).astype(dtype=np.uint8)
    # collapse the three channels
    img_diff = np.max(img_diff,axis=-1)
    # save fake segmentation mask
    im_lab2 = Image.fromarray(img_diff)
    im_lab2.save(path_fake+files_fake[i].split('\\')[-1], "PNG")
    
    # save real segmentation mask (all zeros because there is nothing fake)
    im_real_zero = np.zeros((224,224),dtype=np.uint8)    
    im_real_zero = Image.fromarray(im_real_zero)
    im_real_zero.save(path_real+files_real[i].split('\\')[-1], "PNG")
```
